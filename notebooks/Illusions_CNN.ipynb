```python
# Original fig
plt.rcParams.update({'font.size': 11})
id_il = 54
rgb = 2
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(5,5))
ax.imshow(illusions[id_il,:,:]) # Ilussion = id x pixels in 'y' x pixels in 'x'
x = np.array(range(25, 95)); y = np.ones(len(x))*64
ax.plot(x, y, linewidth=2, color='red')
# First marker
marker_id_1 = 15 
etiqueta = "#1"
ax.scatter(x[marker_id_1],y[marker_id_1], c='k')
ax.annotate(etiqueta, (x[marker_id_1],y[marker_id_1]), textcoords="offset points",xytext=(0,10),ha="center")
# Second marker
marker_id_2 = 60 
etiqueta = "#2"
ax.scatter(x[marker_id_2],y[marker_id_2], c='k')
ax.annotate(etiqueta, (x[marker_id_2],y[marker_id_2]), textcoords="offset points",xytext=(0,10),ha="center")
ax.set_yticklabels([])
ax.set_xticklabels([])
plt.tight_layout()
fig.savefig('Results/White_orig.png')
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(5,5))
y = illusions[id_il,64,25:95][:,rgb] # Green value
# print(len(y))
ax.plot(y, linewidth=2, color='green')
# First marker
marker_id_1 = 15 
etiqueta = "{:.1f}".format(y[marker_id_1])
ax.scatter(marker_id_1,y[marker_id_1], c='k')
ax.annotate(etiqueta, (marker_id_1,y[marker_id_1]), textcoords="offset points",xytext=(0,10),ha="center",fontweight="bold")
# Second marker
marker_id_2 = 60 
etiqueta = "{:.1f}".format(y[marker_id_2])
ax.scatter(marker_id_2,y[marker_id_2], c='k')
ax.annotate(etiqueta, (marker_id_2,y[marker_id_2]), textcoords="offset points",xytext=(0,10),ha="center",fontweight="bold")
# Plot ROIs
ax.plot([11,11],[-0.2,1.2],'r--')
ax.plot([19,19],[-0.2,1.2],'r--')
ax.plot([56,56],[-0.2,1.2],'r--')
ax.plot([64,64],[-0.2,1.2],'r--')
ax.set_ylim([-0.05,1.05])
plt.tight_layout()
plt.savefig('Results/White_orig_line_plots.png')
plt.show()
```
